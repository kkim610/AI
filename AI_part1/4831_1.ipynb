```python
f= open('4831.txt')
#T = int(input())
T=int(f.readline())


for test_case in range(1, T + 1):
    K, N, M = map(int, f.readline().split())       

    L = list(map(int, f.readline().split()))
    print("K N M", K, N, M)
    print("L",L)

    count=0
    D=0
    while D < N:
        flag=False
        for i in range(D+K, D, -1):
            if i in L:
                count += 1
                D = i

                flag = True
                break
        if D >= N-K: break                
        if flag == False:
            count = 0
            break
    print('#%d %d' %(test_case, count))
```

```python
# Dynamic Programming (My Original)


f= open('4831.txt')
#T = int(input())
T=int(f.readline())


for test_case in range(1, T + 1):
    K, N, M = map(int, f.readline().split())       

    L = list(map(int, f.readline().split()))
    #print("K N M", K, N, M)
    #print("L",L)


    C=[float('inf') for _ in range(N+1)]

    for i in range(K+1):
        C[i]=0

    #print(C)
    for i in range(K+1,N+1):
        temp=float('inf')
        for k in range(1,K+1):
            m=i-k
            #print("m",m)
            if m >=0 and m in L:
                #print("m",m)
                V=C[m]+1
                #print("V",V)
                if V < temp:
                    temp=V
        C[i]=temp
        #print("i",i, "C[i]",C[i])
    
    if C[N] == float('inf'):
        result = 0
    else:
        result = C[N]
    print('#%d %d' %(test_case, result))              
```
